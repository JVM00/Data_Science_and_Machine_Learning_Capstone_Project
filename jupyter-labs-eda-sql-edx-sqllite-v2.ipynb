### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
%load_ext sql

# Pick a safe style that exists in most PrettyTable versions
try:
    import prettytable as pt
    for candidate in ("PLAIN_COLUMNS", "MARKDOWN", "DEFAULT"):
        if hasattr(pt, candidate):
            %config SqlMagic.style = candidate
            print("Using PrettyTable style:", candidate)
            break
    else:
        print("No expected PrettyTable styles found. Consider Option C below.")
except Exception as e:
    print("PrettyTable not importable:", e)

# Optional: return DataFrames (bypasses PrettyTable formatting for output)
%config SqlMagic.autopandas = True

# Reconnect and run your query
%sql sqlite:///my_data1.db
%sql SELECT * FROM SPACEXTABLE;
```

```python
# Task 1: Display unique launch sites
# Uses the existing SQL magic and database connection
%sql sqlite:///my_data1.db
%sql SELECT DISTINCT "Launch_Site" FROM SPACEXTABLE ORDER BY "Launch_Site";
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'KSC' 


```python
# Task 2: Display 5 records where launch sites begin with 'KSC'
%sql sqlite:///my_data1.db
%sql SELECT * FROM SPACEXTABLE WHERE "Launch_Site" LIKE 'KSC%' LIMIT 5;
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
# Task 3: Total payload mass carried by boosters launched by NASA (CRS)
%sql sqlite:///my_data1.db
%sql SELECT SUM("PAYLOAD_MASS__KG_") AS total_payload_mass_kg \
FROM SPACEXTABLE \
WHERE "Customer" = 'NASA (CRS)';
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
# Task 4: Average payload mass for booster version F9 v1.1
%sql sqlite:///my_data1.db
%sql SELECT AVG("PAYLOAD_MASS__KG_") AS average_payload_mass_kg \
FROM SPACEXTABLE \
WHERE "Booster_Version" LIKE 'F9 v1.1%';
```

### Task 5

##### List the date where the succesful landing outcome in drone ship was acheived.


_Hint:Use min function_ 


```python
# Task 5: earliest date of successful landing on drone ship
%sql sqlite:///my_data1.db
%sql SELECT MIN("Date") AS earliest_drone_ship_success_date \
FROM SPACEXTABLE \
WHERE "Landing_Outcome" = 'Success (drone ship)';
```

### Task 6

##### List the names of the boosters which have success in ground pad  and have payload mass greater than 4000 but less than 6000


```python
# Task 6: List boosters with successful ground pad landings and payload mass between 4000 and 6000 kg
%sql sqlite:///my_data1.db
%sql SELECT DISTINCT "Booster_Version" AS booster_version \
FROM SPACEXTABLE \
WHERE "Landing_Outcome" LIKE '%ground pad%' \
    AND "PAYLOAD_MASS__KG_" > 4000 \
    AND "PAYLOAD_MASS__KG_" < 6000 \
ORDER BY booster_version;
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
%%sql sqlite:///my_data1.db

SELECT
    CASE
        WHEN "Mission_Outcome" LIKE '%Success%' THEN 'Success'
        WHEN "Mission_Outcome" LIKE '%Fail%' OR "Mission_Outcome" LIKE '%Failure%' THEN 'Failure'
        ELSE 'Other'
    END AS outcome,
    COUNT(*) AS total
FROM SPACEXTABLE
WHERE "Mission_Outcome" LIKE '%Success%' OR "Mission_Outcome" LIKE '%Fail%'
GROUP BY outcome;
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass. Use a subquery.


```python
%%sql sqlite:///my_data1.db

SELECT Booster_Version, Payload_Mass__kg_
FROM SPACEXTABLE
WHERE Payload_Mass__kg_ = (
    SELECT MAX(Payload_Mass__kg_)
    FROM SPACEXTABLE
);
```

### Task 9


##### List the records which will display the month names, succesful landing_outcomes in ground pad ,booster versions, launch_site for the months in year 2017
**Note: SQLLite does not support monthnames. So you need to use substr(Date,6,2) for month, substr(Date,9,2) for date, substr(Date,0,5),='2017' for year.**


```python
%%sql sqlite:///my_data1.db

SELECT
    CASE substr(Date, 6, 2)
        WHEN '01' THEN 'January'
        WHEN '02' THEN 'February'
        WHEN '03' THEN 'March'
        WHEN '04' THEN 'April'
        WHEN '05' THEN 'May'
        WHEN '06' THEN 'June'
        WHEN '07' THEN 'July'
        WHEN '08' THEN 'August'
        WHEN '09' THEN 'September'
        WHEN '10' THEN 'October'
        WHEN '11' THEN 'November'
        WHEN '12' THEN 'December'
    END AS Month_Name,
    Booster_Version,
    Launch_Site,
    Landing_Outcome
FROM SPACEXTABLE
WHERE
    substr(Date, 1, 4) = '2017'
    AND Landing_Outcome LIKE '%Success (ground pad)%';
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python

%%sql sqlite:///my_data1.db

SELECT
    Landing_Outcome,
    COUNT(*) AS Outcome_Count
FROM SPACEXTABLE
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY Landing_Outcome
ORDER BY Outcome_Count DESC;
```
